```python
import pandas as pd
df = pd.read_csv('survey2018.csv')
df.head(5)
```

```python
df['LanguageWorkedWith'].head(5)
```

```python
languages = df['LanguageWorkedWith'] \
    .str.split(';', expand=True)
languages.head(5)
```

```python
summary = languages.apply(
    pd.Series.value_counts)
summary.head(10)
```

```python
summary = pd.DataFrame({'count': 
    summary.sum(axis=1)
    .groupby(lambda x: x.strip())
    .sum()})
summary
```

```python
total = df[df['LanguageWorkedWith'].notnull()].shape[0]
summary['percent'] = summary['count']/total*100
summary
```

```python
summary.plot.barh(y='percent', figsize=(10,10))
```
